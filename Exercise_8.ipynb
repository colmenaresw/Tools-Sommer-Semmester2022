```python
from sympy import *
from sympy.plotting import plot

x = symbols('x')  # we define x as a symbol

# defining the function
f_x = 2*x + 8 + (66/(x-8))

# plotting
p1 = plot(f_x, (x,0,4), line_color='r', title="polynomial function",  show=True)
p1 = plot(f_x, (x,0,20), line_color='y', title="polynomial function", ylim=(-100,100),  show=True)

zeros = solve(f_x, x)  # we calculate the zeros of the function

# generation of intervals
i_zeros = Interval(zeros[0], zeros[1])  # we create an interval with the zeros of the function
i = Interval(0, 5)

# other operations
deriv = diff(f_x, x)  # we calculate its derivative
maxi = maximum(f_x, x, i)  # we calculate the maxima inside the interval [0, 5], otherwise it is infinity
integral_def = integrate(f_x, (x, zeros[0], zeros[1]))  # we calculate the integral inside the interval

# printing of results
print(f'RESULTS:\n \
        zeros: {zeros}\n \
        derivative: {deriv}\n \
        integral: {integral_def.evalf()}\n \
        maximum: {maxi}\n \
      ')
```
